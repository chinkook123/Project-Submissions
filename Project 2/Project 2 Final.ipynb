## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. Explain your rationale for the way you choose to "impute" the missing data.

```python
df1 = df.dtypes # Displays the all the types of all the columns in the dataset
df1
```

```python
df2 = df[['x1st.week']] = df[['x1st.week']].astype(float) # converts the 'x1st.week' column 
# from a int to a float
df2.dtypes
```

```python
pd.pivot_table(df, index = ['artist.inverted', 'track', 'genre'], values = ['x1st.week']).sort()
# Create pivot table
```

```python
dfWeeksOnBillboard = df.notnull().sum(axis = 1)
dfWeeksOnBillboard
```

```python
remove_na = df.dropna(axis=1, how = 'all') # Drops the columns that contain all 
# NaN values (weeks 66-76)
remove_na
```

```python
remove_na.count() # Counts the number of values in columns, NaN values don't show up
```

```python
column_names = remove_na.columns.tolist()
column_names[column_names.index('year')] = 'Year'
column_names[column_names.index('artist.inverted')] = 'Artist'
column_names[column_names.index('track')] = 'Track'
column_names[column_names.index('time')] = 'Time'
column_names[column_names.index('genre')] = 'Genre'
column_names[column_names.index('date.entered')] = 'Date Entered'
column_names[column_names.index('date.peaked')] = 'Date Peaked'

for i in column_names:
    y = i  
    if str(i).startswith('x'):
        i = i[1: ]
        #print(i)
        i = str(i).replace('.', ' ')
        #print(i)
        column_names[column_names.index ( y) ] = i
        remove_na.columns = column_names
print (column_names)
```

```python
piv_tab = pd.pivot_table(remove_na, index = ('Artist', 'Track'))
piv_tab
```

```python
# find out the number of weeks each song has been on the billboard
dfWeeksOnBillboard1 = piv_tab.notnull().sum(axis = 1)
dfWeeksOnBillboard1
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
melted = pd.melt(piv_tab.reset_index(), id_vars=['Artist', 'Track'],
            var_name='col', value_name='val')
melted
```

```python
billboard_top = melted[melted.val == 1].sort_values(by='Artist')
billboard_top
```

```python
top_songs = billboard_top['Track'].value_counts()
top_songs
```

```python
billboard_top10 = melted[melted.val <= 10].sort_values(by='Artist')
billboard_top10
```

```python
top10_songs = billboard_top10['Track'].value_counts()
top10_songs
```
